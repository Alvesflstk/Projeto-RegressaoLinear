```
Importando As bibliotecas necessárias
```



````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import uniform
````



```
Entrada De Dados
```



````python
planilha = pd.read_excel('/content/drive/MyDrive/Despesas (2).xlsx')
````



```
Classe Despesas
```



````python
class despesas_gerais:
  def __init__(self,alimentacao,transporte,verificacoes,meses):
    self.alimentacao = alimentacao 
    self.transporte = transporte 
    self.verificacoes = verificacoes
    self.meses = meses  

geral = despesas_gerais(planilha['Alimentação'].values,planilha['Transporte'].values,planilha['Verificações'].values,planilha['Meses'].values)

````



```
Gráfico de Co-relação De Dados 
```



````python
plt.plot(geral.meses,geral.alimentacao,label='Alimentação')
plt.scatter(geral.meses,geral.alimentacao)
plt.plot(geral.meses,geral.transporte,label = 'Transporte')
plt.scatter(geral.meses,geral.transporte)
plt.plot(geral.meses,planilha['Verificações'], label = 'Verificações')
plt.scatter(geral.meses,geral.verificacoes,color = 'tomato')
plt.legend()
plt.show()
````

***`class de gráficos`***

````python
class Graficos:
  def hipotese(w0,w1,x):
    return w0+w1*x
````

````python
Graficos.hipotese(ws[0], ws[1],x)
````

````python
plt.scatter(x,y,color='b')
plt.plot(x,Graficos.hipotese(ws[50],ws[51],x),color = 'r')
````
